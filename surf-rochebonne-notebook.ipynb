```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# Links

url = 'https://www.windguru.cz/500910'
executable_path = '/Users/marionchaffaut/chromedriver/chromedriver-mac-x64/chromedriver'
```

```python
# Init scraping

driver = webdriver.Chrome(executable_path=executable_path)
driver.get(url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
# Saves 

text_file = open("forecast.txt", "w")
text_file.write(str(soup.find_all('script')))      
text_file.close()
```

```python
# Find key features

dates = soup.find("table", {"class": "tabulka"}).find("tbody").find_all("tr", {"id": "tabid_0_0_dates"})
wind_speed = soup.find("table", {"class": "tabulka"}).find("tbody").find_all("tr", {"id": "tabid_0_0_WINDSPD"})
wind_gust = soup.find("table", {"class": "tabulka"}).find("tbody").find_all("tr", {"id": "tabid_0_0_GUST"})
swell_height = soup.find("table", {"class": "tabulka"}).find("tbody").find_all("tr", {"id": "tabid_0_0_HTSGW"})
swell_period = soup.find("table", {"class": "tabulka"}).find("tbody").find_all("tr", {"id": "tabid_0_0_PERPW"})
```

```python
# Extract key features

def process_items(items, forecast):
    for item in items:
        cells = item.find_all("td")
        id = item['id']
        forecast[id] = []
        for i, cell in enumerate(cells):
            if 'DIRPW' in id:
                print(f"{id} {i}")
                value = cell.find('span').find('svg').find('g')["transform"]
            else:
                value = cell.get_text()
            forecast[id].append(value)

forecast = {}

features = [dates, wind_speed, wind_gust, swell_height, swell_period]

for items in features:
    process_items(items, forecast)

forecast_df = pd.DataFrame(forecast)
```
